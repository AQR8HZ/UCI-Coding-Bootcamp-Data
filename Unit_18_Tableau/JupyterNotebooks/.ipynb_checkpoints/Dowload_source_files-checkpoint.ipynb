# This Code downloads all available files for the TRIPDATA analysis
* The code identifies the CSV files URL's and downloads every single file
* Files are downloads as ZIP files and extracted as CSV as a prerequisite for the ETL process

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import requests, zipfile, io
```

```python
 # URL of page to be scraped
url = 'https://s3.amazonaws.com/tripdata'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')

# Select all files identified with <key>
files = soup.find_all('key')
```

```python
# Download and Extract each selected file
for file in files:
    file_url = f"{url}/{file.text}"
    print (file_url)
    r = requests.get(file_url)
    z = zipfile.ZipFile(io.BytesIO(r.content))
    z.extractall('Data/source')
```
