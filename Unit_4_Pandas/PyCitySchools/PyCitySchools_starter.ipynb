```python
# Run basic calculations to obtain number of unique items, average price, etc
total_Schools = len(school_data_complete["School ID"].unique())
total_Students = len(school_data_complete["Student ID"].unique())
total_Budget = (school_data_complete.groupby(['School ID'])["budget"].max()).sum()
average_Math_Score = round(school_data_complete["math_score"].mean(),2)
average_Reading_Score = round(school_data_complete["reading_score"].mean(),2)
overall_Passing_Score = round((average_Math_Score + average_Reading_Score)/2,2)

perc_Students_Math_Score_GT69 = ((len(school_data_complete.loc[school_data_complete['math_score'] >= 70]))/total_Students) * 100
perc_Students_Reading_Score_GT69 = ((len(school_data_complete.loc[school_data_complete['reading_score'] >= 70]))/total_Students) * 100

# Create a summary data frame to hold the results
schoolSummary_df = pd.DataFrame([
    {"Total Schools": total_Schools,
     "Total Students": total_Students,
     "Total Budget": total_Budget,
     "Average Math Score": average_Math_Score,
     "Average Reading Score": average_Reading_Score,
     "% Passing Math": perc_Students_Math_Score_GT69,
     "% Passing Reading": perc_Students_Reading_Score_GT69,
     "% Overall Passing Rate": overall_Passing_Score}
])

# Organize columns to create the desired outcome
organized_df = schoolSummary_df[[
    "Total Schools",
     "Total Students",
     "Total Budget",
     "Average Math Score",
     "Average Reading Score",
     "% Passing Math",
     "% Passing Reading",
     "% Overall Passing Rate"    
]]

# Optional: give the displayed data cleaner formatting
# Use Map to format Budget($), and %'s columns
organized_df["Total Budget"] = organized_df["Total Budget"].map("${:,.2f}".format)
organized_df["% Passing Math"] = organized_df["% Passing Math"].map("{:.2f}".format)
organized_df["% Passing Reading"] = organized_df["% Passing Reading"].map("{:.2f}".format)

#Display the summary data frame
organized_df.head()
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# Calculate the total count by gender, avoiding duplicated players
school_Type           = school_data_complete.groupby(['school_name'])['type'].max()
total_School_Students = school_data_complete.groupby(['school_name'])['Student ID'].count()
total_School_Budget   = school_data_complete.groupby(['school_name'])['budget'].max()
per_Student_Budget    = total_School_Budget / total_School_Students
average_Math_Score    = round(school_data_complete.groupby(['school_name'])['math_score'].mean(),3)
average_Reading_Score = round(school_data_complete.groupby(['school_name'])['reading_score'].mean(),3)

passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70, ['school_name', 'math_score']]
passing_math_bySchool = passing_math.groupby(['school_name'])['math_score'].count()
perc_passing_math_bySchool = (passing_math_bySchool / total_School_Students) * 100

passing_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70, ['school_name', 'reading_score']]
passing_reading_bySchool = passing_reading.groupby(['school_name'])['reading_score'].count()
perc_passing_reading_bySchool = (passing_reading_bySchool / total_School_Students) * 100

overall_Passing_Score = (perc_passing_math_bySchool + perc_passing_reading_bySchool)/2


# Create the data frame for presenting the data
top_PerformingSchools = pd.DataFrame({
    'School Type'            : school_Type,
    'Total Students'         : total_School_Students,
    'Total School Budget'    : total_School_Budget,
    'Per Student Budget'     : per_Student_Budget,
    'Average Math Score'     : average_Math_Score,
    'Average Reading Score'  : average_Reading_Score,
    '% Passing Math'         : perc_passing_math_bySchool,
    '% Passing Reading'      : perc_passing_reading_bySchool,
    '% Overall Passing Rate' : overall_Passing_Score
})


# Use Map to format Budget($), and %'s columns
top_PerformingSchools["Total School Budget"] = top_PerformingSchools["Total School Budget"].map("${:,.2f}".format)
top_PerformingSchools["Per Student Budget"] = top_PerformingSchools["Per Student Budget"].map("${:,.2f}".format)
top_PerformingSchools["% Passing Math"] = top_PerformingSchools["% Passing Math"].map("{:.3f}".format)
top_PerformingSchools["% Passing Reading"] = top_PerformingSchools["% Passing Reading"].map("{:.3f}".format)
top_PerformingSchools["% Overall Passing Rate"] = top_PerformingSchools["% Overall Passing Rate"].map("{:.3f}".format)

# Sort the data in decending order to match the required output
top_PerformingSchools.sort_values(by=['% Overall Passing Rate'], ascending=False).head(5)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

analysis_By = 'Spending Ranges (Per Student)'
new_Column_Name = 'Per Student Budget' # column added to the source table

# Calculate student population by school and budget per student
total_School_Students = school_data_complete.groupby(['school_name'])['Student ID'].count()
total_School_Budget   = school_data_complete.groupby(['school_name'])['budget'].max()


# Same code as previous section
# ----------------------------------------------
# This section should be implemented as a function for code simplification purposes, need to nvestigate more for Jupyter N.



new_Column = pd.DataFrame({new_Column_Name : (total_School_Budget / total_School_Students)})

# Merge the budget per student to the shool data table and create a new table "Merge School Table"
merge_school_table = pd.merge(school_data_complete, new_Column, on='school_name')

# Assign the bins to the source data as a new column
merge_school_table[analysis_By] = pd.cut(merge_school_table[new_Column_Name], bins, labels=group_names)


# Calculate the data by bin using groupby
# Total students by bin range
total_Students_byRange = merge_school_table.groupby([analysis_By])['Student ID'].count()

# Calculate Average Math and Reading scores by bin range
average_Math_Score = merge_school_table.groupby([analysis_By])['math_score'].mean()
average_Reading_Score = merge_school_table.groupby([analysis_By])['reading_score'].mean()

# Calculate passing rates for Math, Reading and Overall scores by bin range
passing_math = merge_school_table.loc[merge_school_table['math_score'] >= 70, [analysis_By, 'math_score']]
passing_math_byRange = passing_math.groupby([analysis_By])['math_score'].count()
perc_passing_math_byRange = (passing_math_byRange / total_Students_byRange) * 100

passing_reading = merge_school_table.loc[merge_school_table['reading_score'] >= 70, [analysis_By, 'reading_score']]
passing_reading_byRange = passing_reading.groupby([analysis_By])['reading_score'].count()
perc_passing_reading_byRange = (passing_reading_byRange / total_Students_byRange) * 100

overall_Passing_Score_byRange = (perc_passing_math_byRange + perc_passing_reading_byRange)/2

# Create the data frame for presenting the data
scores_Df = pd.DataFrame({
    'Average Math Score'     : average_Math_Score,
    'Average Reading Score'  : average_Reading_Score,
    '% Passing Math'         : perc_passing_math_byRange,
    '% Passing Reading'      : perc_passing_reading_byRange,
    '% Overall Passing Rate' : overall_Passing_Score_byRange
})

# Use Map to format %'s columns
scores_Df["Average Math Score"]     = scores_Df["Average Math Score"].map("{:,.3f}".format)
scores_Df["Average Reading Score"]  = scores_Df["Average Reading Score"].map("{:,.3f}".format)
scores_Df["% Passing Math"]         = scores_Df["% Passing Math"].map("{:.3f}".format)
scores_Df["% Passing Reading"]      = scores_Df["% Passing Reading"].map("{:.3f}".format)
scores_Df["% Overall Passing Rate"] = scores_Df["% Overall Passing Rate"].map("{:.3f}".format)

scores_Df
```
